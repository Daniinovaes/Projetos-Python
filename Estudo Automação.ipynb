```python
shoppings_nome = []
shoppings_url = []
paginas = [1,2,3,4,5,6,7]
```

```python
import requests
from bs4 import BeautifulSoup
import re

for pagina in paginas:
    r = requests.get('https://abrasce.com.br/guia-de-shoppings/page/'+str(pagina)+'/')
    html = r.text
    
    soup = BeautifulSoup(html,'html.parser')
    
    for shoppings in soup.find_all("div", attrs={"class":"shopping"}):
        shopping_url = shoppings.a['href']
        shoppings_url.append(shopping_url)
        shop = shoppings.a.find("div", attrs={"class":"row"}).text
        shop = shop.strip()
        shoppings_nome.append(shop)
```

```python
shop = []
```

```python
from selenium import webdriver
import time

# delay
segundos = 5

for site in shoppings_url:  
    info_shopping = []
    # nova sessão no Chrome
    navegador = webdriver.Safari()
    
    # navegar na pagina da abrasce
    navegador.get(site)
    
    time.sleep(segundos)
    html_n = navegador.page_source
    
    soup_n = BeautifulSoup(html_n, "html.parser")
    
    nome_shopping = soup_n.h1.text
    perfil_a = soup_n.find("span", attrs={"id":"mfs_perfil_a"}).text
    perfil_b = soup_n.find("span", attrs={"id":"mfs_perfil_b"}).text
    perfil_c = soup_n.find("span", attrs={"id":"mfs_perfil_c"}).text
    perfil_d = soup_n.find("span", attrs={"id":"mfs_perfil_d"}).text
    area_construida = soup_n.find("span", attrs={"id":"mfs_contruida"}).text
    area_total = soup_n.find("span", attrs={"id":"mfs_total"}).text
    abl_shop = soup_n.find("span", attrs={"id":"mfs_bruta"}).text
    entretenimento = soup_n.find("span", attrs={"id":"mfs_entretenimento"}).text
    endereco = soup_n.find("span", attrs={"id":"mfs_endereco"}).text
    pisos = soup_n.find("p", attrs={"id":"mfs_piso_loja"}).text
    ancora = soup_n.find("p", attrs={"id":"mfs_loja_ancora"}).text
    lojas = soup_n.find("p", attrs={"id":"mfs_total_lojas"}).text
    cinema = soup_n.find("p", attrs={"id":"mfs_cinema"}).text
    estacionamento = soup_n.find("p", attrs={"id":"mfs_vagas"}).text
    
    
    info_shopping.append(nome_shopping)
    info_shopping.append(endereco)
    info_shopping.append(area_total)
    info_shopping.append(area_construida)
    info_shopping.append(abl_shop)
    info_shopping.append(perfil_a)
    info_shopping.append(perfil_b)
    info_shopping.append(perfil_c)
    info_shopping.append(perfil_d)
    info_shopping.append(entretenimento)
    info_shopping.append(pisos)
    info_shopping.append(ancora)
    info_shopping.append(lojas)
    info_shopping.append(cinema)
    info_shopping.append(estacionamento)
    
    shop.append(info_shopping)

    navegador.quit()
```

```python
import pandas as pd

shop_df = pd.DataFrame(shop, columns = ['Nome', 'Endereco', 'Área Total', 'Área Contruida', 'ABL', 'Perfil A (%)', 'Perfil B (%)', 'Perfil C (%)', 'Perfil D (%)', 'Entretenimento', 'Pisos', 'Qtde Lojas Âncora', 'Qtde Lojas', 'Salas Cinema', 'Vagas Estacionamento'])
```
